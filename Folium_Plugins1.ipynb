# Plugin HeatMap



```python
# Zuerst die grundlegende Karte erstellen
import folium

map_swiss = folium.Map(
    location=[47.36635101042978, 8.541361139590903],
    zoom_start=13,
    attr="© <a href='https://www.swisstopo.admin.ch/'>swisstopo</a>",
    control_scale=True,
    tiles="https://wmts.geo.admin.ch/1.0.0/ch.swisstopo.swissimage/default/current/3857/{z}/{x}/{y}.jpeg"
)

```

Bevor die weiteren Schritt ausgeführt werden, müssen die Geo-Daten hochgeladen werden!

```python
# Vector-Layer als GeoJson einlesen
import json

f = open("./stzh_bezirke_plgn.json", 'r')
zh_bez = json.load(f)
```

```python
# Karte mit GeoJson anzeigen
folium.features.GeoJson(zh_bez, name="Stadt Zürich").add_to(map_swiss)

map_swiss
```

```python
import pandas as pd
import ast
from folium import plugins

wc_df = pd.read_csv("stzh_poi_zueriwc_att.csv")

# Daten für Heatmap erstellen
wc_arr = []
for x in wc_df["coord"]:
    tp = ast.literal_eval(x)
    wc_point = [tp[0], tp[1]]
    wc_arr.append(wc_point)

folium.plugins.HeatMap(data=wc_arr).add_to(map_swiss)
map_swiss
```
