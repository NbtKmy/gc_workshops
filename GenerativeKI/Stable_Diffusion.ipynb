```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16)
```

```python
pipe = pipe.to("cuda")
```

## Argumente für StableDiffusionPipeline

StableDiffusionPipeline kann nicht nur "prompt" sondern auch andere Argumente wie "negative prompt" nehmen.

Welche Argumente dort möglich sind, sieht man [hier](https://huggingface.co/docs/diffusers/v0.20.0/en/api/pipelines/stable_diffusion/text2img#diffusers.StableDiffusionPipeline.__call__)



```python
prompt = "steampunk illustration of an anthromorphic cat as detective in victorian fashion, costume highly detailed, english city in the background, dark and horror atmosphere, high quality, monotone"
image = pipe(prompt).images[0]


image
```
