# TEI?

TEI (Text Encoding Initiative) steht sowohl für den Standard, mit dem man diverse digitale Texte maschninenlesbar strukturiert, als auch die Organisation, die den Standard pflegt.

Mittlerweile sind viele Texte nach TEI codiert und mit der offenen Nutzungslizenzen publiziert.

Bevor wir konkret daran basteln, lernen wir grob über TEI.

Die Beschreibung von TEI als Standard regel für Text-Markup ist in [Guidline](https://tei-c.org/release/doc/tei-p5-doc/en/html/index.html) zu finden. Dort sind viele Elemente und Attribute genau definiert, unter welchen Situation sie eingesetzt werden sollen. Durch die Regelung können wir erwarten, dass die unterschiedlichen TEI-codierten Dokumente ähnliche Struktur haben. - So erleichtert TEI maschinelle Behandlung der Texte.

Die aktuelle Version (P5, Stand: 22.08.2023) basiert auf XML. Im deutschsprachigen Wikipedia-Artikel findet man ein Paar kurze Beispiele:

## Quiz 1

Wir haben jetzt die Struktur des Dokuments kennengelernt.
Wie können wir die Worte von Macbeth daraus holen?

````python
from bs4 import BeautifulSoup

with open("macbeth.xml") as f:
    data = f.read()

soup = BeautifulSoup(data, "lxml-xml")
````

````python
# Z.B. mit find_all - Wir wollen alle Inhalte mit <sp>-Tags und who="#Macbeth_Mac" holen
macbeth = soup.find_all("sp", attrs = {"who": "#Macbeth_Mac"})
print(macbeth)
````

````python
# Unter sp-Tags sind <l>-Tags (Line) vorhanden. Diese l-Tag tragen die Worte von Macbeth

def get_dialog_from_sp(sp):
    lines = sp.find_all("l")
    line_string = ""

    for l in lines:
        ws = l.find_all("w")
        for w in ws:
            line_string += w.get_text() + " "

    return line_string

all_macbeth_speech = ""
for sp in macbeth:
    lines = get_dialog_from_sp(sp)
    all_macbeth_speech += lines + "\n"

print(all_macbeth_speech)
````

Überlegen wir weiter für den Fall, wenn wir Wordcloud von Macbeths Rede erstellen möchte.

Zum Glück ist hat jede `<w>`-Tags "lemma"-Attribut, woher wir Lemmata holen können. In diesem Beispiel nehmen wir nur die Nomen aus dem Text.


````python
!pip install -q wordcloud
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def get_lemmata_from_sp(sp):
    lines = sp.find_all("l")
    line_array = []
    for l in lines:
        ws = l.find_all("w")
        for w in ws:
            w_att = w.attrs
            # Es gibt anscheinend `w`-Tags, die `ana`-Attribute nicht besitzen...
            if "ana" not in w_att:
                continue
            if "#n" in w_att["ana"]:
                #print(w_att["lemma"])
                line_array.append(w_att["lemma"])

    return line_array

all_macbeth_speech_array = []
for sp in macbeth:
    lemmata_lines = get_lemmata_from_sp(sp)
    all_macbeth_speech_array += lemmata_lines


# Genaue beschreibung für Wordcloud : https://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html#wordcloud.WordCloud
wordcloud = WordCloud(
                height=400
            )
wc = wordcloud.generate(" ".join(all_macbeth_speech_array))
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
````
